```python
import os 
ROOT_DIR = os.path.dirname(os.path.abspath(os.getcwd()))
ROOT_DIR
```

```python
DATA_FOLDER_PATH = os.path.join(ROOT_DIR, 'data')
os.makedirs(DATA_FOLDER_PATH, exist_ok=True)
```

```python
# CDS API configuration script
import cdsapi
from dotenv import load_dotenv

load_dotenv()

URL = os.getenv("URL")
API_KEY = os.getenv("API_KEY")

home_path = os.path.expanduser('~')
config_file = os.path.join(home_path, '.cdsapirc')

with open(config_file, 'w', encoding='utf-8') as f:
    f.write(f'url: {URL}\n')
    f.write(f'key: {API_KEY}\n')
print(f'--- File .cdsapirc path: {config_file} ---')

try:
    c = cdsapi.Client()
    print("--- Connect Sucess ---")
except Exception as e:
    print(f"Error: {e}")
```

```python
import yaml 
YAML_FILE = os.path.join(ROOT_DIR, 'config.yaml')

try:
    with open(YAML_FILE, 'r', encoding='utf-8') as f:
        config = yaml.safe_load(f)
except Exception as e:
    print(f"Error loading YAML file: {e}")

for key in config:
    print(f"--- Config key: {key} ---")
```

```python
try:
    print(f"Downloading data to folder: {DATA_FOLDER_PATH}")
    c.retrieve(
        'reanalysis-era5-single-levels',
        config['2m_temperature_downloader'],
        os.path.join(DATA_FOLDER_PATH, '2m_temperature.nc')
    )
    print("--- 2m_temperature download completed ---")
except Exception as e:
    print(f"Error downloading 2m_temperature: {e}")
```

```python
try:
    c.retrieve(
        'reanalysis-era5-single-levels',
        config['wind_10m_downloader'],
        os.path.join(DATA_FOLDER_PATH, 'wind_10m.nc')
    )
    print("--- wind_10m download completed ---")
except Exception as e:
    print(f"Error downloading wind_10m: {e}")
```
